```python
# Import libraries
from skimage import io, img_as_float, img_as_ubyte
import numpy as np
from skimage.filters import gaussian
img1 = img_as_float(io.imread("Image.jpg", as_gray=True))
```

```python
# Apply Gaussian filter on the image
gaussian_img = gaussian(img1, sigma=1, mode='constant', cval=0.0, multichannel=False)
```

```python
# Perfrom Unsharp Masking
img2 = (img1 - gaussian_img)*1.
img3 = img1 + img2
```

```python
# Plot the result
from matplotlib import pyplot as plt
plt.imshow(img3, cmap='gray')
```

```python
# Convery the range of pixels to [0, 255]
a=min(img3.flatten())
b=max(img3.flatten())
def remap(x, in_min, in_max, out_min, out_max):
  return (x - in_min) * (out_max - out_min) / (in_max - in_min) + out_min
new=remap(img3, a, b, 0, 255)
```

```python
# In numpy, the values are always rounded down (e.g., 121.90 --> 121).
# To make it round to the nearest integer, the following should be done. 
new=new+0.5
new=new.astype(np.uint8) #Integer Values
```

```python
plt.imshow(new, cmap='gray')
```
